```python
from OCR.text_detector import detect
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# read image
img = cv2.imread('Files/facture5.jpg')
```

```python
# Convert to gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# renforce contours
thresh = 170
ret, thresh_img = cv2.threshold(img_gray, thresh, 255, cv2.THRESH_BINARY)
```

```python
# apply the text_detector methods
detections = detect(thresh_img,
                  kernel_size=101,  # valeur initiale 25 doit erte un nombre impair
                  sigma=40,
                  theta=5,
                  min_area=2000)
```

```python
# draw the detections and plot the results
for cord_boxes in detections:
    cords = cord_boxes.bbox
    cv2.rectangle(img, (int(cords.x), int(cords.y)), (int(cords.x + cords.w), int(cords.y + cords.h)), (0, 0, 255), 2)

plt.figure(figsize = (25,17))
plt.imshow(img)
plt.show()
```
