```python
from IPython.core.display import HTML
from matplotlib import pyplot as plt
import deepdoctection as dd
from pathlib import Path
```

```python
analyzer = dd.get_dd_analyzer()  # instantiate the built-in analyzer similar to the Hugging Face space demo

df = analyzer.analyze(path = "Files/facture5.pdf")  # setting up pipeline
df.reset_state()                 # Trigger some initialization

doc = iter(df)
page = next(doc)
```

```python
image = page.viz()
plt.figure(figsize = (25,17))
plt.axis('off')
plt.imshow(image)
```

```python
print(page.text)
```

```python
HTML(page.html)
```

```python
for layout in page.layouts:
    if layout.category_name=="title":
        print(f"Title: {layout.text}")
    elif layout.category_name=="header":
        print(f"Header: {layout.text}")
    elif layout.category_name=="footer":
        print(f"Footer: {layout.text}")
    elif layout.category_name=="table":
        print(f"Table: {layout.text}")
    elif layout.category_name=="paragraph":
        print(f"Paragraph: {layout.text}")
    elif layout.category_name=="list":
        print(f"List: {layout.text}")
    elif layout.category_name=="figure":
        print(f"Figure: {layout.text}")
    elif layout.category_name=="other":
        print(f"Other: {layout.text}")
    else:
        print(f"Unknown: {layout.text}")
```
